```python
def get_song_page(num, year):
    if num < 100:
        url = "https://en.wikipedia.org/wiki/Billboard_year-end_top_" + str(num) + "_singles_of_" + str(year)
    else:
        url = "https://en.wikipedia.org/wiki/Billboard_Year-End_Hot_100_singles_of_" + str(year)
    page = requests.get(url)
    # It doesn't work
    if (page.status_code != 200):
        if (num == 30):
            return get_song_page(50, year)
        elif (num == 50):
            return get_song_page(100, year)
    # It works!
    else:
        #print(year)
        get_citation_page(page)
        return page
```

```python
def get_citation_page(page):
    soup = BeautifulSoup(page.content, "html.parser")
    for link in soup.find_all("a"):
        if link.get("title") == "Information on how to cite this page":
            url = "https://en.wikipedia.org"+link.get("href")
            break        
    page = requests.get(url)
    if (page.status_code != 200):
        print("FAILED!")
        return
    else:
        return get_citation(page)
```
