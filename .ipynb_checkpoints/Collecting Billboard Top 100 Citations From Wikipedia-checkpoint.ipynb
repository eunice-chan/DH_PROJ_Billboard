```python
import requests
import csv
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
```

```python
def get_page(num, year):
    if num < 100:
        url = "https://en.wikipedia.org/wiki/Billboard_year-end_top_" + str(num) + "_singles_of_" + str(year)
    else:
        url = "https://en.wikipedia.org/wiki/Billboard_Year-End_Hot_100_singles_of_" + str(year)
    page = requests.get(url)
    if (page.status_code != 200):
        if (num == 30):
            return get_page(50, year)
        elif (num == 50):
            return get_page(100, year)
    else:
        print(year)
        return page
```

```python
#Collecting lists of the Billboard Top 100 from the years 1950-2018 from Wikipedia
#Dropped 50th place tie Oh Lonesome Me in 1958
begin = 1950
end = 2018
years = np.arange(begin, end+1) #range of dates available: 1950-2018
pos = []
song = []
artist = []
yr = []
print("Finished collecting Billboard's Top 100 for the years:")
for year in years:
    page = get_page(30, year)
    if (page.status_code == 200):
        soup = BeautifulSoup(page.content, "html.parser")
        tb = soup.find("table", class_="wikitable").find("tbody") #top 100 (may be less) list stored in wikitable
        if tb:
            for row in tb.find_all("tr")[1:]:
                row = [ elem for elem in row.text.split("\n") if elem]
                if len(row) == 3:
                    pos += [row[0]]
                    song += [row[1]]
                    artist += [row[2]]
                    yr += [year]
                else:
                    print("Dropped", row)
```

```python
d = {"Position": pos, "Song": song, "Artist": artist, "Year": yr}
billboard = pd.DataFrame(data=d)
billboard.head()
```

```python
billboard.to_csv("Billboard100_"+str(begin)+"-"+str(end)+".csv",index=False)
```
